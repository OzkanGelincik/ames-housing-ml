#  <h1 style="text-align:center; color:yellow; font-size:72px;">**Machine Learning Project - Ames Housing Data**</h1>


Ames, Iowa is the college town of **Iowa State University**. The Ames housing dataset consists of about $2500$ house sale records between $2006-2010$. Detailed information about the house attributes, along with the sale prices, is recorded in the dataset. The goal of the project is to:
- perform descriptive data analysis to gain business (i.e. housing market) insights
- build descriptive machine learning models to understand the local housing market.
- build predictive machine learning models for the local house price prediction.

A subset of the **Ames** dataset is hosted on [**Kaggle**](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data) as an entry-level regression competition. You may visit their site for some information on the meanings of its data columns (the data dictionary). In this notebook, we will describe various project ideas related to this data.

# Feature Extractor

#### Here I define a feature extraction function that systematically checks a model's attributes to identify the object, peel off any wrappers, grab the actual model, extract the coefficients or feature importances, construct a dataframe ready to for the next function to plot. 

```python
# ==========================================
# UNIVERSAL EXTRACTION FUNCTION
# ==========================================

def extract_model_features(model_object, X_train_columns):
    """
    Extracts feature names and weights (coefficients or importances) from a pipeline.
    Handles TransformedTargetRegressor, CorrelationThreshold, and different model types.
    """
    # 1. Drill down to the Pipeline
    if hasattr(model_object, 'regressor_'):
        # It's a TransformedTargetRegressor
        pipeline = model_object.regressor_
    elif hasattr(model_object, 'best_estimator_'):
        # It's a GridSearchCV object
        pipeline = model_object.best_estimator_.regressor_
    else:
        pipeline = model_object

    # 2. Get the current list of feature names (before the final model sees them)
    # We start with all columns and simulate the drops
    current_features = np.array(X_train_columns)
    
    if 'corr_filter' in pipeline.named_steps:
        corr_step = pipeline.named_steps['corr_filter']
        dropped_cols = set(corr_step.to_drop_)
        # Keep only features NOT in the drop list
        current_features = np.array([f for f in current_features if f not in dropped_cols])

    # 3. Get the Final Model Step
    final_step_name = list(pipeline.named_steps.keys())[-1]
    final_model = pipeline.named_steps[final_step_name]
    
    # 4. Extract Weights based on Model Type
    df = pd.DataFrame()
    
    # CASE A: Linear Models (Lasso, Ridge, ElasticNet, OLS) -> Coefficients
    if hasattr(final_model, 'coef_'):
        weights = final_model.coef_
        metric_name = "Coefficient"
        
        # Lasso/ElasticNet specific: Filter out exact zeros
        mask = np.abs(weights) > 1e-5
        current_features = current_features[mask]
        weights = weights[mask]
        
    # CASE B: Tree Models (RandomForest, XGBoost) -> Feature Importances
    elif hasattr(final_model, 'feature_importances_'):
        weights = final_model.feature_importances_
        metric_name = "Importance"
        
    # CASE C: CatBoost (Has its own method) -> Feature Importances
    elif hasattr(final_model, 'get_feature_importance'):
        weights = final_model.get_feature_importance()
        metric_name = "Importance"
        
    # CASE D: Voting Regressor (Ensemble) -> Cannot plot single set of weights
    elif hasattr(final_model, 'estimators_'):
        print("Note: VotingRegressor has no single set of weights. Visualize base models individually.")
        return pd.DataFrame() # Return empty
        
    else:
        print(f"Unknown model type: {type(final_model)}")
        return pd.DataFrame()

    # 5. Build DataFrame
    df = pd.DataFrame({
        'Feature': current_features,
        metric_name: weights
    })
    
    # Add absolute value column for sorting
    df['Abs_Weight'] = df[metric_name].abs()
    return df.sort_values(by='Abs_Weight', ascending=False)
```

# Feature Impact Visuzlization Function

#### Here I defined a plotting function that takes in the dataframe containing a model's coefficients or feature importance, sorts the top 20 features by the absolute values of their coefficients or importances (i.e., weights), then plots them as horizontal bar plot.

```python
# ==========================================
# FEATURE IMPACT VISUALIZATION FUNCTION
# ==========================================

def plot_feature_impact(df, title="Feature Impact"):
    if df.empty:
        print("No data to plot.")
        return

    # Determine metric type (Coefficient or Importance)
    metric_col = [c for c in df.columns if c in ['Coefficient', 'Importance']][0]
    
    # Take Top 20 for readability
    plot_df = df.head(20).sort_values(by='Abs_Weight', ascending=True) # Ascending for horiz bar
    
    plt.figure(figsize=(10, 8))
    
    # Color Logic
    if metric_col == 'Coefficient':
        colors = ['green' if x > 0 else 'red' for x in plot_df[metric_col]]
        xlabel = "Impact on Log-Price (Green=Positive, Red=Negative)"
    else:
        colors = 'skyblue'
        xlabel = "Feature Importance (0 to 1)"

    plt.barh(plot_df['Feature'], plot_df[metric_col], color=colors)
    
    # Aesthetics
    if metric_col == 'Coefficient':
        plt.axvline(x=0, color='black', linestyle='--', linewidth=0.8)
        
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Features")
    plt.tight_layout()
    plt.show()
```

## `GridSearchCV()` for Linear Models (Lasso, Ridge, ElasticNet) *(with OHE only, i.e., `X_train_ohe` and `X_test_ohe`)*

```python
# ==========================================
# 1. RE-DEFINE PIPELINES (Standardizing Step Names)
# ==========================================
# We define them here to ensure the inner step is named 'model'
# This matches the param_grids below.

# Lasso
lasso_pipe = Pipeline([
    ('corr_filter', CorrelationThreshold(threshold=0.9)),
    ('scaler', StandardScaler()),
    ('model', Lasso(random_state=42)) # Step name is 'model'
])
lasso_full = TransformedTargetRegressor(regressor=lasso_pipe, func=np.log1p, inverse_func=np.expm1)

# Ridge
ridge_pipe = Pipeline([
    ('corr_filter', CorrelationThreshold(threshold=0.9)),
    ('scaler', StandardScaler()),
    ('model', Ridge(random_state=42)) # Step name is 'model'
])
ridge_full = TransformedTargetRegressor(regressor=ridge_pipe, func=np.log1p, inverse_func=np.expm1)

# ElasticNet
elastic_pipe = Pipeline([
    ('corr_filter', CorrelationThreshold(threshold=0.9)),
    ('scaler', StandardScaler()),
    ('model', ElasticNet(random_state=42, max_iter=50000)) # Step name is 'model'
])
elastic_full = TransformedTargetRegressor(regressor=elastic_pipe, func=np.log1p, inverse_func=np.expm1)

# ==========================================
# 2. SETUP PARAMETER GRIDS
# ==========================================
# Now these keys are guaranteed to work because we forced the name 'model' above.

param_grid_lasso = {
    'regressor__model__alpha': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1]
}

param_grid_ridge = {
    'regressor__model__alpha': [0.1, 1.0, 5.0, 10.0, 50.0, 100.0, 200.0]
}

param_grid_elastic = {
    'regressor__model__alpha': [0.0001, 0.001, 0.01, 0.1],
    'regressor__model__l1_ratio': [0.1, 0.5, 0.7, 0.9, 0.99] 
}

# ==========================================
# 3. THE TUNING LOOP
# ==========================================
models_to_tune = [
    ('Lasso', lasso_full, param_grid_lasso),
    ('Ridge', ridge_full, param_grid_ridge),
    ('ElasticNet', elastic_full, param_grid_elastic)
]

results = []

print("Starting Grid Search Exploration...")
print("-----------------------------------")

for name, model, grid in models_to_tune:
    print(f"Tuning {name}...")
    
    search = GridSearchCV(
        model, 
        grid, 
        cv=5, 
        scoring='r2', 
        n_jobs=-1,
        verbose=1
    )
    
    # Fit on the "Dirty" OHE data
    search.fit(X_train_ohe, y_train)
    
    # Capture results
    best_score = search.best_score_
    best_params = search.best_params_
    
    # Clean up params for display
    clean_params = {k.split('__')[-1]: v for k, v in best_params.items()}
    
    results.append({
        'Model': name,
        'Best CV Score': best_score,
        'Best Params': clean_params
    })
    
    print(f"  -> Best CV Score: {best_score:.4f}")
    print(f"  -> Best Params: {clean_params}\n")

# ==========================================
# 4. FINAL LEADERBOARD
# ==========================================
results_df = pd.DataFrame(results).sort_values(by='Best CV Score', ascending=False)

print("\n--- FINAL LINEAR MODEL LEADERBOARD ---")
pd.set_option('display.max_colwidth', None)
print(results_df)
```

## `GridSearchCV()` for SVR Model *(with OHE only, i.e., `X_train_ohe` and `X_test_ohe`)*

### SVR with rbf kernel

```python
# 1. Define SVR Pipeline
# We use the same structure: Filter -> Scale -> SVR
svr_pipe = Pipeline([
    ('corr_filter', CorrelationThreshold(threshold=0.9)),
    ('scaler', StandardScaler()),
    ('model', SVR(kernel='rbf', cache_size=2000))
])

svr_full = TransformedTargetRegressor(
    regressor=svr_pipe, 
    func=np.log1p, 
    inverse_func=np.expm1
)

# 2. Define Parameter Grid
# SVR is VERY sensitive to these three:
param_grid_svr = {
    # C: Penalty parameter. High C = strict fit (good for OHE). Low C = smooth fit.
    'regressor__model__C': [1, 10, 100, 1000], 
    
    # epsilon: The "tube" size. Smaller = fit closer to the line.
    'regressor__model__epsilon': [0.01, 0.1, 0.2],
    
    # gamma: Defines how far the influence of a single training example reaches.
    # 'scale' is default, but 'auto' or specific numbers often help OHE data.
    'regressor__model__gamma': ['scale', 0.001, 0.01, 0.1]
}

print("Tuning SVR on OHE Data (This may take 5-10 minutes)...")

search_svr = GridSearchCV(
    svr_full, 
    param_grid_svr, 
    cv=3,                 # Lower CV to 3 to save time (SVR is slow)
    scoring='r2', 
    n_jobs=-1,
    verbose=1
)

search_svr.fit(X_train_ohe, y_train)

# 3. Results
print(f"Best SVR Score: {search_svr.best_score_:.4f}")
print(f"Best Params: {search_svr.best_params_}")
```

## `GridSearchCV` for Tree Models *(with ordinal only, i.e., `X_train_ordinal` and `X_test_ordinal`)*

```python
# ==========================================
# 1. RE-DEFINE PIPELINES (Standardized for Trees)
# ==========================================
# Note: Tree models do NOT need Scaling or Correlation Filtering.
# We strip those out to make them faster and more accurate.

# Random Forest
rf_pipe = Pipeline([
    ('model', RandomForestRegressor(random_state=42)) # Step name is 'model'
])
rf_full = TransformedTargetRegressor(regressor=rf_pipe, func=np.log1p, inverse_func=np.expm1)

# XGBoost
xgb_pipe = Pipeline([
    ('model', XGBRegressor(random_state=42, n_jobs=-1)) # Step name is 'model'
])
xgb_full = TransformedTargetRegressor(regressor=xgb_pipe, func=np.log1p, inverse_func=np.expm1)

# CatBoost (Verbose=0 to silence output)
cb_pipe = Pipeline([
    ('model', CatBoostRegressor(random_state=42, verbose=0)) # Step name is 'model'
])
cb_full = TransformedTargetRegressor(regressor=cb_pipe, func=np.log1p, inverse_func=np.expm1)

# ==========================================
# 2. SETUP PARAMETER GRIDS
# ==========================================
# Critical: The key path is 'regressor' -> 'model' -> 'parameter'

# Random Forest
param_grid_rf = {
    'regressor__model__n_estimators': [100, 300],
    'regressor__model__max_depth': [None, 10, 20],
    'regressor__model__min_samples_leaf': [1, 3] # Prevents overfitting
}

# XGBoost
param_grid_xgb = {
    'regressor__model__learning_rate': [0.01, 0.05, 0.1],
    'regressor__model__max_depth': [3, 5, 7],
    'regressor__model__n_estimators': [500, 1000],
    'regressor__model__subsample': [0.8, 1.0]
}

# CatBoost
param_grid_cb = {
    'regressor__model__depth': [4, 6, 8],
    'regressor__model__learning_rate': [0.01, 0.05, 0.1],
    'regressor__model__l2_leaf_reg': [1, 3, 5]
}

# ==========================================
# 3. THE TUNING LOOP
# ==========================================
models_to_tune = [
    ('RandomForest', rf_full, param_grid_rf),
    ('XGBoost', xgb_full, param_grid_xgb),
    ('CatBoost', cb_full, param_grid_cb)
]

results = []

print("Starting Tree Model Grid Search...")
print("-----------------------------------")

for name, model, grid in models_to_tune:
    print(f"Tuning {name}...")
    
    # n_jobs=-1 uses all cores. cv=3 is fast enough for trees.
    search = GridSearchCV(
        model, 
        grid, 
        cv=3, 
        scoring='r2', 
        n_jobs=-1,
        verbose=1
    )
    
    # CRITICAL: Fit on ORDINAL data for trees
    search.fit(X_train_ordinal, y_train)
    
    best_score = search.best_score_
    best_params = search.best_params_
    
    # Clean parameter names for display
    clean_params = {k.split('__')[-1]: v for k, v in best_params.items()}
    
    results.append({
        'Model': name,
        'Best CV Score': best_score,
        'Best Params': clean_params
    })
    
    print(f"  -> Best CV Score: {best_score:.4f}")
    print(f"  -> Best Params: {clean_params}\n")

# ==========================================
# 4. FINAL LEADERBOARD
# ==========================================
results_df = pd.DataFrame(results).sort_values(by='Best CV Score', ascending=False)

print("\n--- FINAL TREE MODEL LEADERBOARD ---")
pd.set_option('display.max_colwidth', None)
print(results_df[['Model', 'Best CV Score', 'Best Params']])
```

## Voting Regressor (Ensemble) - Tuned

```python
# ======================================================
# 1. DEFINE INDIVIDUAL BRANCHES
# ======================================================

# --- Branch A: Lasso ---
# FIX: Removed 'CorrelationThreshold'. 
# Lasso handles multicollinearity natively, and this fixes the "numpy has no .drop()" error.
n_cats = len(categorical_cols)

lasso_pipe = Pipeline([
    # Uses ColumnTransformer to target ONLY categorical columns
    ('prep', ColumnTransformer([
        ('ohe', OneHotEncoder(categories='auto', sparse_output=False, handle_unknown='ignore'), slice(0, n_cats))
    ], remainder='passthrough')),
    ('scaler', StandardScaler()),
    ('model', Lasso(alpha=0.001, max_iter=50000, random_state=42)) # Updated alpha
])

# --- Branch B: XGBoost ---
# Updated with your new best params
xgb_model = XGBRegressor(
    n_estimators=500,       # Updated
    learning_rate=0.1,      # Updated
    max_depth=3,            # Updated
    subsample=0.8,          # Updated
    random_state=42,
    n_jobs=-1               # Let XGBoost use all cores internally
)

# --- Branch C: CatBoost ---
# Updated with your new best params
cb_model = CatBoostRegressor(
    iterations=1000,        # Kept at 1000 (standard safe upper limit)
    learning_rate=0.05,     # Updated
    depth=4,                # Updated
    l2_leaf_reg=3,          # Updated
    loss_function='RMSE',
    random_seed=42,
    verbose=0,
    allow_writing_files=False
)

# ======================================================
# 2. DEFINE THE VOTING ENSEMBLE
# ======================================================
# FIX: Changed n_jobs to 1. 
# Since XGB and CatBoost use all cores, n_jobs=-1 here causes CPU fighting/deadlocks.
voting_model = VotingRegressor(
    estimators=[
        ('lasso', lasso_pipe),
        ('xgb', xgb_model),
        ('catboost', cb_model)
    ],
    weights=[1, 2, 2], 
    n_jobs=1 
)

# ======================================================
# 3. WRAP IN LOG-TRANSFORMER
# ======================================================
final_ensemble_model = TransformedTargetRegressor(
    regressor=voting_model,
    func=np.log1p,
    inverse_func=np.expm1
)

# ======================================================
# 4. TRAIN AND EVALUATE
# ======================================================
print("Training Final Ensemble (Lasso + XGBoost + CatBoost)...")

# This should now run much faster than 15 minutes
final_ensemble_model.fit(X_train_ordinal, y_train)
print(f"Train R^2: {final_ensemble_model.score(X_train_ordinal, y_train):.5f}")

# Score
score = final_ensemble_model.score(X_test_ordinal, y_test)
print(f"\n--- Final Ensemble Diagnostics ---")
print(f"Test R^2: {score:.5f}")

# Comparison
best_single_model = 0.914489 # Your previous best
if score > best_single_model:
    print(f"🎉 SUCCESS! Ensemble beat the single best model by {score - best_single_model:.5f}")
else:
    print(f"Result: {score - best_single_model:.5f} difference from best single model.")


final_score = final_ensemble_model.score(X_test_ordinal, y_test)
final_preds = final_ensemble_model.predict(X_test_ordinal)
final_mae = mean_absolute_error(y_test, final_preds)
final_rmse = root_mean_squared_error(y_test, final_preds)
final_rmsle = root_mean_squared_log_error(y_test, final_preds)

print(f"\n--- Final Ensemble Metrics ---")
print(f"Test R^2: {final_score:.5f}")
print(f"MAE:      ${final_mae:,.0f}")
print(f"RMSE:     ${final_rmse:,.0f}")
print(f"RMSLE:    ${final_rmsle:,.5f}")
```

```python
def predict_safe(model, input_data):
    """
    Wraps the model prediction with safety guardrails.
    Prevents the 'Lasso Catapult' effect on extreme outliers.
    """
    # 1. Create a copy so we don't modify the original
    safe_data = input_data.copy()
    
    # 2. Define Safety Limits (Based on your training data max values)
    MAX_SQ_FT = 10000  # Cap at 10k sq ft
    MAX_QUAL = 10      # Cap quality at 10
    
    # 3. Clamp the values
    if 'GrLivArea' in safe_data.columns:
        # If user types 200,000, we force it down to 10,000
        safe_data['GrLivArea'] = safe_data['GrLivArea'].clip(upper=MAX_SQ_FT)
        
    if 'OverallQual' in safe_data.columns:
        safe_data['OverallQual'] = safe_data['OverallQual'].clip(upper=MAX_QUAL)

    # 4. Predict
    prediction = model.predict(safe_data)[0]
    return prediction

# --- Re-Test the Mega Mansion ---
print(f"--- Re-Testing Mega Mansion with Guardrails ---")
safe_price = predict_safe(final_ensemble_model, mansion_house)
print(f"Safe Prediction: ${safe_price:,.2f}")
```

## Extracting Precise Neighborhood Coordinates

```python
# 1. Setup the Geocoder
# User_agent is required so Nominatim knows who is asking
geolocator = Nominatim(user_agent="ames_housing_project_v1")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# 2. Define the Ames Neighborhood Mapping
# The dataset uses abbreviations. We need to map them to real search queries.
# "Ames, IA" is appended to ensure we don't find a "North Ames" in Texas.
neighborhood_map = {
    'CollgCr': 'College Creek, Ames, IA',
    'Veenker': 'Veenker, Ames, IA',
    'Crawfor': 'Crawford, Ames, IA',
    'NoRidge': 'Northridge, Ames, IA',
    'Mitchel': 'Mitchell, Ames, IA',
    'Somerst': 'Somerset, Ames, IA',
    'NWAmes': 'Northwest Ames, Ames, IA',
    'OldTown': 'Old Town, Ames, IA',
    'BrkSide': 'Brookside, Ames, IA',
    'Sawyer': 'Sawyer, Ames, IA',
    'NridgHt': 'Northridge Heights, Ames, IA',
    'NAmes': 'North Ames, Ames, IA',
    'SawyerW': 'Sawyer West, Ames, IA',
    'IDOTRR': 'Iowa DOT and Rail Road, Ames, IA',
    'MeadowV': 'Meadow Village, Ames, IA',
    'Edwards': 'Edwards, Ames, IA',
    'Timber': 'Timberland, Ames, IA',
    'Gilbert': 'Gilbert, Ames, IA',
    'StoneBr': 'Stone Brook, Ames, IA',
    'ClearCr': 'Clear Creek, Ames, IA',
    'NPkVill': 'Northpark Village, Ames, IA',
    'Blmngtn': 'Bloomington Heights, Ames, IA',
    'BrDale': 'Briardale, Ames, IA',
    'SWISU': 'South & West of ISU, Ames, IA',
    'Blueste': 'Bluestem, Ames, IA'
}

# 3. Create the Coordinate Dictionary
# This runs once and saves results so we don't re-query the API.
neighborhood_coords = {}

print("Starting Geocoding (this will take ~30 seconds)...")

for code, search_query in neighborhood_map.items():
    try:
        location = geocode(search_query)
        if location:
            neighborhood_coords[code] = (location.latitude, location.longitude)
            print(f"✅ Found: {code} -> {location.latitude}, {location.longitude}")
        else:
            # Fallback for hard-to-find places (Manual approximate center of Ames)
            print(f"⚠️ Could not find: {code}. Using default.")
            neighborhood_coords[code] = (42.0347, -93.6200) 
    except Exception as e:
        print(f"❌ Error for {code}: {e}")
        neighborhood_coords[code] = (42.0347, -93.6200)

print("Geocoding Complete.")
```

```python
# ==========================================
# 1. DEFINE YOUR RENOVATION MENU
# ==========================================
# This dictionary holds the "Recipe" for each renovation.
# You can add new items here easily!

renovations_menu = {
    "Kitchen Upgrade (Granite/Modern)": {
        "changes": {"KitchenQual": "Gd"}, 
        "cost_retail": 15000, 
        "cost_investor": 8000  # DIY/Wholesale materials
    },
    "Add Central Air": {
        "changes": {"CentralAir": "Y"}, 
        "cost_retail": 6000, 
        "cost_investor": 3500
    },
    "Finish Basement (500 sqft)": {
        "changes": {"BsmtFinSF1": 500, "TotalBsmtSF": 500}, 
        "cost_retail": 25000, 
        "cost_investor": 12000
    },
    "Add 2-Car Garage": {
        "changes": {"GarageCars": 2, "GarageArea": 500}, 
        "cost_retail": 30000, 
        "cost_investor": 18000
    },
    "Full Rehab (Kitchen + AC + Bsmt)": {
        "changes": {
            "KitchenQual": "Gd", 
            "CentralAir": "Y", 
            "BsmtFinSF1": 500,
            "TotalBsmtSF": 500
        },
        "cost_retail": 46000,   # Sum of individual retail costs
        "cost_investor": 23500  # Sum of individual investor costs
    }
}
```

### The "Simulation"

```python
# ==========================================
# 2. RUN THE SIMULATION (Dual Scenarios)
# ==========================================
simulation_results = []

# Get Base Price
base_price = production_model.predict(base_house)[0]

print(f"🏠 Base House Value: ${base_price:,.2f}\n")
print("-" * 95)
print(f"{'RENOVATION':<35} | {'VALUE ADD':<12} | {'RETAIL ROI':<12} | {'INVESTOR ROI':<12}")
print("-" * 95)

for name, params in renovations_menu.items():
    renovated_house = base_house.copy()

    # Apply changes
    for feature, new_value in params['changes'].items():
        if feature in ['BsmtFinSF1', 'TotalBsmtSF']:
             if feature == 'TotalBsmtSF' and renovated_house[feature].iloc[0] == 0:
                 renovated_house[feature] = new_value 
             else:
                 renovated_house[feature] += new_value
        else:
            renovated_house[feature] = new_value

    # Predict New Price
    new_price = production_model.predict(renovated_house)[0]
    value_lift = new_price - base_price

    # --- SCENARIO 1: RETAIL (Homeowner) ---
    cost_retail = params['cost_retail']
    roi_retail = ((value_lift - cost_retail) / cost_retail) * 100

    # --- SCENARIO 2: INVESTOR (Flipper/DIY) ---
    cost_investor = params['cost_investor']
    roi_investor = ((value_lift - cost_investor) / cost_investor) * 100

    # Store Data
    simulation_results.append({
        "Renovation": name,
        "Value Lift": value_lift,
        "Retail Cost": cost_retail,
        "Retail ROI": roi_retail,
        "Investor Cost": cost_investor,
        "Investor ROI": roi_investor,
        "New Price": new_price
    })

    # Print Row
    print(f"{name:<35} | ${value_lift:,.0f}      | {roi_retail:>6.1f}%      | {roi_investor:>8.1f}%")

# ==========================================
# 3. DISPLAY FINAL TABLE
# ==========================================
results_df = pd.DataFrame(simulation_results)

# Display with nice formatting
display(results_df.style.format({
    "Value Lift": "${:,.0f}",
    "Retail Cost": "${:,.0f}",
    "Retail ROI": "{:.1f}%",
    "Investor Cost": "${:,.0f}",
    "Investor ROI": "{:.1f}%",
    "New Price": "${:,.0f}"
}).background_gradient(subset=["Retail ROI", "Investor ROI"], cmap="RdYlGn"))
```
